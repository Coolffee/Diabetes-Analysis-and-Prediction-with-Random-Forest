```python
Healthy =  df[ df['Diabetes_binary'] == 0 ]
Diabetic = df[ df['Diabetes_binary'] == 1 ]

h_diab = pd.Series({'Saudável':Healthy.shape[0],
            'Diabetes':Diabetic.shape[0]})
h_diab.plot.bar(alpha=0.7)
```

```python
# porcentagem de mulheres com diabetes
h_diab.plot.pie(startangle=90,
                explode=[0, 0.1],
                autopct='%1.1f%%',
                colors=['C3', 'C4'])
plt.title(' ')
plt.ylabel('')
_ = plt.axis('equal')
```

```python
#Proporção de Casos de Diabetes entre Homens e Mulheres

diabetes_men = df[(df['Diabetes_binary'] == 1) & (df['Sex'] == 0)]
diabetes_women = df[(df['Diabetes_binary'] == 1) & (df['Sex'] == 1)]

# Contar o número de casos de diabetes para homens e mulheres
total_diabetes_men = len(diabetes_men)
total_diabetes_women = len(diabetes_women)

# Plotar o gráfico de pizza
labels = ['Homens', 'Mulheres']
sizes = [total_diabetes_men, total_diabetes_women]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # Destacar a fatia de "Homens"

plt.figure(figsize=(5, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=['C3', 'C4'], autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('')
plt.show()
```

```python
# Calcular a contagem de casos de diabetes por idade para homens e mulheres
age_diabetes = df.groupby('Age')['Diabetes_binary'].mean()

# Plotar o gráfico de barras em roxo
plt.bar(age_diabetes.index, age_diabetes.values, color='purple')
plt.xlabel('Idade')
plt.ylabel('Prevalência de Diabetes')
plt.title('Prevalência de Diabetes por Idade')
plt.show()
```

```python
# criando X e y
X=df.drop('Diabetes_binary',axis=1)
y=df['Diabetes_binary']
```

```python
# separando em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaling_x=StandardScaler()
X_train=scaling_x.fit_transform(X_train)
X_test=scaling_x.transform(X_test)
```

```python
# rede com 100 arvores
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# metricas do modelo
from sklearn.metrics import *
y_pred = model.predict(X_test)
acurracy = accuracy_score(y_test,y_pred)
report = classification_report(y_test,y_pred)

print(f'Acuracia: {acurracy}\n Classificação: \n{report}')
```

```python
feature_importances = model.feature_importances_

# Obtenha os nomes das features
feature_names = X.columns

# Crie um gráfico de barras para visualizar a importância das features
plt.barh(feature_names, feature_importances)
plt.xlabel('Importância')
plt.title('Importância das Features em um Modelo Random Forest')
plt.show()
```

```python
# rede 2 com 5 arvores
model2 = RandomForestClassifier(n_estimators=5)
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test)
acurracy2 = accuracy_score(y_test,y_pred2)
report2 = classification_report(y_test,y_pred2)

print(f'Acuracia: {acurracy2}\n Classificação: \n{report2}')
```
